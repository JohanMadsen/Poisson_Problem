```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def surface_plot (matrix, **kwargs):
    # acquire the cartesian coordinate matrices from the matrix
    # x is cols, y is rows
    a = 2*(np.arange(matrix.shape[0])/matrix.shape[0]-0.5)
    b = 2*(np.arange(matrix.shape[1])/matrix.shape[1]-0.5)
    (x, y) = np.meshgrid(a,b)
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(x, y, matrix, **kwargs)
    ax.view_init(35, 45+180)
    fig.tight_layout()
    return (fig, ax, surf)
```

```python
m = np.loadtxt("test.txt")
(fig, ax, surf) = surface_plot(m, cmap=plt.cm.RdYlGn_r)

#fig.colorbar(surf)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('U')
ax.set_title("Plot of U on X,Y surface", fontsize=18)

plt.savefig("verify.png", transparant=True)

plt.show()
```
